```python
import numpy as np
import pandas as pd
import seaborn as sns
import re
```

```python
df_fpa = pd.read_csv("fpa/fpa_3mf_result", delimiter='&')
df_fpa = df_fpa[['Name ', ' PI ', ' CI ', ' maxE ']]
df_fpa['Method'] = 'fpa'
df_fpa['Month'] = df_fpa['Name '].map(lambda s: ''.join(re.findall('[A-Z]+',s)))

df_wm = pd.read_csv("wm/wm_3mf_result", delimiter='&')
df_wm = df_wm[['Name ', ' PI ', ' CI ', ' maxE ']]
df_wm['Method'] = 'wm'
df_wm['Month'] = df_wm['Name '].map(lambda s: ''.join(re.findall('[A-Z]+',s)))

df_ols = pd.read_csv("ols/ols_results", delimiter='&')
df_ols = df_ols[['Name ', ' PI ', ' CI ', ' maxE ']]
df_ols['Method'] = 'ols'
df_ols['Month'] = df_ols['Name '].map(lambda s: ''.join(re.findall('[A-Z]+',s)))
```

```python
df = df_fpa.append(df_ols).append(df_wm)
```

```python
sns.set(rc={'figure.figsize':(15, 7)})
ax = sns.barplot(x = "Month", y = " PI ", hue = "Method", data = df)

figure = ax.get_figure()   
figure.tight_layout()
figure.savefig('barplot_PI_fpa-ols-wm.png', dpi=400)
```

```python
sns.set(rc={'figure.figsize':(15, 7)})
ax = sns.barplot(x = "Month", y = " CI ", hue = "Method", data = df)
ax.set(ylim=(0.4, 1))

figure = ax.get_figure()   
figure.tight_layout()
figure.savefig('barplot_CI_fpa-ols-wm.png', dpi=400)
```
